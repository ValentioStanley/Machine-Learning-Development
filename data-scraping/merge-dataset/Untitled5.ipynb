```python
import glob
import os
import pandas as pd

path = '/content/merge-dataset/'
csv_files = glob.glob(os.path.join(path, '*.csv'))

merged_dataset = pd.concat([pd.read_csv(berkas) for berkas in csv_files], ignore_index=True)
merged_dataset.to_csv('/content/main-dataset/dataset-total.csv', index=False)

print(f"{len(csv_files)} CSV files merged successfully.")
```
